## Extracting Date Fields

### Although the date field for any given article is available to be scraped, I found that it was more efficient to omit any date tags from the scraping. Rather, once the article URL was scraped, I wrote a simple function implementing regex logic to parse this URL and return only the date identified in the URL. 

### Once the date has been extracted, the apply function is used to generate a new column in the dataframe containing the correct date value.


```python
import re, datetime
def regex_date(string):
    try:
        match = re.search('\d{4}/\d{2}/\d{2}',string)
        date = datetime.datetime.strptime(match.group(), '%Y/%m/%d').date()
        return date
    except:
        return None
```

```python
result1['url_date'] =result1['url'].apply(regex_date)
```
