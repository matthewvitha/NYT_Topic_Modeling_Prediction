```python
from random import randint
from time import sleep

url = "http://api.nytimes.com/svc/search/v2/articlesearch.json?q='United States Economy'&page={}&begin_date=20100103&sort=newest&api-key=34ae6398b29445188f15963b07e30352"
def nyt_get(url):
    #begin_date=20100103 is the date of half the guardian articles
    nyt_list = []

        
    for i in range(1,200):

        #url = url.format(i)
        r = requests.get(url.format(i))
        sleep(randint(1,2))
        print(url.format(i))
        print(r.status_code)
        if r.status_code == 200:
            
            nyt_list.append(r.json())
            
        else:
            print('cant pull no more :(')

    return nyt_list
```
