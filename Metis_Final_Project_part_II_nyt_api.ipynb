# Metis Final Project Part II. Gathering Articles from the NYT API from 1982 to 2000.


### The below code continues to gather article meta data from the NYT API.  

```python
import requests
from pprint import pprint
```

```python
from random import randint
from time import sleep

url10 = "http://api.nytimes.com/svc/search/v2/articlesearch.json?q='United States Economy'&page={}&begin_date=19930101&end_date=20000517&sort=newest&api-key=34ae6398b29445188f15963b07e30352"
def nyt_get10(url):
    #begin_date=20100103 is the date of half the guardian articles-from before not accurate here
    nyt_list = []

        
    for i in range(1,200):

        #url = url.format(i)
        r = requests.get(url.format(i))
        sleep(randint(1,2))
        print(url.format(i))
        print(r.status_code)
        if r.status_code == 200:
            
            nyt_list.append(r.json())
            
        else:
            print('cant pull no more :(')

    return nyt_list
```

```python
nyt_list_11 = nyt_get10(url10)
```

```python
nyt_list_11[198]
```

```python
import pickle
with open('nyt_list_11.pickle', 'wb') as f11:
    pickle.dump(nyt_list_11, f11)
```
